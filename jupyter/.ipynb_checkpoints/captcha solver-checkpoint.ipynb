```python
from math import sqrt
from skimage import io, feature, filters
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
im = io.imread('captchas/captcha1.png', as_gray=True)
```

```python
plt.imshow(im, cmap="gray")
plt.show()
```

```python
val = filters.threshold_otsu(im)
drops = ndimage.binary_fill_holes(im > val)
drops = ndimage.binary_hit_or_miss(im > val)
plt.imshow(drops, cmap='gray')
plt.show()
```

```python
from skimage.filters import try_all_threshold


fig, ax = try_all_threshold(im, figsize=(10, 8), verbose=False)
plt.show()
```

```python
from skimage.filters import threshold_mean


thresh = threshold_mean(im)
binary = im < thresh

fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
ax = axes.ravel()

ax[0].imshow(im, cmap=plt.cm.gray)
ax[0].set_title('Original image')

ax[1].imshow(binary, cmap=plt.cm.gray)
ax[1].set_title('Result')

for a in ax:
    a.axis('off')

plt.show()
```

```python
blobs_log = feature.blob_log(im, max_sigma=30, num_sigma=10, threshold=.15)
blobs_log[:, 2] = blobs_log[:, 2] * 3

blobs_dog = feature.blob_dog(im, max_sigma=30, threshold=.1)
blobs_dog[:, 2] = blobs_dog[:, 2] * 2

blobs_doh = feature.blob_doh(im, max_sigma=30, threshold=.01)

# assign into array
blobs_list = [blobs_log, blobs_dog, blobs_doh]

colors = ["green", "yellow", "red"]

titles = ["Laplacian of Gaussian", "Difference of Gaussian", "Determinent of Hessian"]

sequence = zip(blobs_list, colors, titles)

# validate whethere we capture all the dots

fig, axes = plt.subplots(1, 3, figsize=(15, 10), sharex=True, sharey=True)
ax = axes.ravel()

for idx, (blobs, color, title) in enumerate(sequence):
    numstar = len(blobs)
    ax[idx].set_title(f"{title} [Num of star: {numstar}]")
    ax[idx].imshow(im)
    for blob in blobs:
        y, x, r = blob
        c = plt.Circle((x, y), r, color=color, linewidth=2, fill=False)
        ax[idx].add_patch(c)
    ax[idx].set_axis_off()

plt.tight_layout()
plt.show()
```
