```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
image = cv2.imread("captchas/captcha7.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
canny = cv2.Canny(gray, 28, 70, L2gradient=True)
plt.imshow(canny, cmap="gray")
```

```python
dilated = cv2.dilate(canny, (2, 2), iterations=2)
plt.imshow(dilated, cmap="gray")
```

```python
cnt, heirarchy = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print("star counts: ", len(cnt))
# rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
# plt.imshow(rgb)
```

```python
fig, axes = plt.subplots(20, figsize=(30, 50), sharex=True, sharey=True)
ax = axes.ravel()

for idx in range(20):
    low = randint(20, 35)
    while True:
        heigh = randint(40, 55)
        if heigh > low:
            break
    
    ax[idx].set_title(f"Low: {low} heigh: {heigh}")
    ax[idx].imshow(cv2.Canny(gray, low, heigh, L2gradient=True), cmap="gray")
    ax[idx].set_axis_off()

# plt.tight_layout()
plt.show()
```
